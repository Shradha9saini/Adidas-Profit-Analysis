```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
Asd = pd.read_csv('AdidasSalesdata-csv.csv') 

print(Asd.head())
```

```python
# Data Cleaning 
Asd.isnull().sum()
Asd.info()
Asd['Invoice Date']= pd.to_datetime(Asd['Invoice Date'])
Asd['Units Sold'] = Asd['Units Sold'] * 10
Asd['Year']= Asd['Invoice Date'].dt.strftime('%Y')
Asd['M/Y']=Asd['Invoice Date'].dt.strftime('%Y/%m')
Asd.head()
```

```python
#Year on year profit

Asd_2020 =Asd[Asd['Year'] == '2020']
Asd_2021 = Asd[Asd['Year'] == '2021']

Profit20= Asd_2020['Operating Profit'].sum()
Profit21= Asd_2021['Operating Profit'].sum()
YonYprofit= round((Profit21 - Profit20),2)
print("adidas increased their operating profits for 2020 to 2021 by: " + str(YonYprofit))
```

```python
# Percentage increase in profit

pcincrease=round((YonYprofit/Profit20)*100,2)
print(str(pcincrease) + '%')
```

```python
#Distribution by Retailer

Profit_retailer = Asd.groupby('Retailer').sum('Operating Profit').sort_values(by = 'Operating Profit', ascending= False)
Profit_retailer['Operating Profit'].plot(kind='bar').set_ylabel('Operating Profit (MM)')
print(Profit_retailer["Operating Profit"])



```

```python
#Profits by retailer per year


Asd_2020['Operating Profit'].groupby(Asd['Retailer']).sum('Operating Profit').sort_values(ascending = False).plot(kind= 'bar').set_ylabel('Operating Profit 2020')

print(Asd_2020['Operating Profit'].groupby(Asd['Retailer']).sum('Operating Profit').sort_values(ascending = False))

  

```

```python
Asd_2021['Operating Profit'].groupby(Asd['Retailer']).sum('Operating Profit').sort_values(ascending = False).plot(kind='bar').set_ylabel('Operating Profit 2021')
print(Asd_2021['Operating Profit'].groupby(Asd['Retailer']).sum('Operating Profit').sort_values(ascending = False))
```

```python
#Percentage increase in profit from 2020-2021 

Asd_increased_profit= (((Asd_2021.groupby('Retailer').sum('Operating Profit')) - (Asd_2020.groupby('Retailer').sum('Operating Profit')))/(Asd_2020.groupby('Retailer').sum('Operating Profit')))*100
Asd_increased_profit['Operating Profit'].sort_values(ascending = False ).plot(kind='bar')
```
